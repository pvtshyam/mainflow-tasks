```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Data Loading
# Load the dataset
file_path = r"C:\Users\dell\Documents\student-mat.csv"  # Replace with the correct file path
data = pd.read_csv(file_path)
print("Dataset loaded successfully!")

# Display the first few rows
print(data.head())

# Task 2: Data Exploration
# Check for missing values
print("\nMissing values in each column:")
print(data.isnull().sum())

# Display column data types
print("\nData types of each column:")
print(data.dtypes)

# Display dataset size
print("\nDataset size (rows, columns):")
print(data.shape)

# Task 3: Data Cleaning
# Handle missing values (if any)
data.fillna(data.median(numeric_only=True), inplace=True)
print("\nMissing values handled.")

# Remove duplicate entries
data.drop_duplicates(inplace=True)
print("\nDuplicates removed.")

# Task 4: Data Analysis
# 1. Average final grade (G3)
average_g3 = data['G3'].mean()
print(f"\n1. Average final grade (G3): {average_g3:.2f}")

# 2. Count of students scoring above 15 in G3
students_above_15 = (data['G3'] > 15).sum()
print(f"2. Number of students scoring above 15 in G3: {students_above_15}")

# 3. Correlation between study time and G3
correlation = data['studytime'].corr(data['G3'])
print(f"3. Correlation between study time and G3: {correlation:.2f}")

# 4. Average final grade by gender
average_g3_by_gender = data.groupby('sex')['G3'].mean()
print("\n4. Average final grade by gender:")
print(average_g3_by_gender)

# Task 5: Data Visualization
# 1. Histogram of final grades (G3)
plt.figure(figsize=(8, 6))
sns.histplot(data['G3'], bins=10, kde=False, color='blue')
plt.title('Histogram of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()

# 2. Scatter plot: study time vs final grade
plt.figure(figsize=(8, 6))
sns.scatterplot(x='studytime', y='G3', data=data, color='green')
plt.title('Study Time vs Final Grade')
plt.xlabel('Study Time (hours/week)')
plt.ylabel('Final Grade (G3)')
plt.show()

# 3. Bar chart: average scores by gender
plt.figure(figsize=(8, 6))
average_g3_by_gender.plot(kind='bar', color=['purple', 'orange'])
plt.title('Average Final Grade by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.xticks(rotation=0)
plt.show()
```
